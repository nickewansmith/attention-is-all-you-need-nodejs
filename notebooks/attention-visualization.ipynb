# Transformer Attention Visualization
Inspect encoder, decoder self, and cross-attention matrices returned by the NestJS Transformer API.

```python
import json
import requests
import numpy as np
import matplotlib.pyplot as plt
```

Configure the API endpoint.

```python
API_URL = 'http://localhost:3000/transformer/translate'
```

```python
payload = {
    'text': 'attention is all you need',
    'maxLength': 32,
    'strategy': 'beam',
    'beamWidth': 4,
    'includeAttention': True
}
response = requests.post(API_URL, json=payload)
response.raise_for_status()
data = response.json()
data
```

Extract attention tensors and visualize them.

```python
attention = data.get('attention')
if isinstance(attention, list):
    attention = attention[0]
encoder_layers = attention['encoder']
decoder_cross_layers = attention['decoderCross']
```

Plot an encoder head.

```python
layer_idx, head_idx = 0, 0
matrix = np.array(encoder_layers[layer_idx]['heads'][head_idx])
plt.imshow(matrix, cmap='viridis')
plt.title(f'Encoder Layer {layer_idx+1}, Head {head_idx+1}')
plt.xlabel('Key Position')
plt.ylabel('Query Position')
plt.colorbar()
plt.show()
```

Plot cross-attention (decoder→encoder alignment).

```python
layer_idx, head_idx = 0, 0
matrix = np.array(decoder_cross_layers[layer_idx]['heads'][head_idx])
plt.imshow(matrix, cmap='magma')
plt.title(f'Cross-Attention Layer {layer_idx+1}, Head {head_idx+1}')
plt.xlabel('Encoder Tokens')
plt.ylabel('Decoder Tokens')
plt.colorbar()
plt.show()
```
